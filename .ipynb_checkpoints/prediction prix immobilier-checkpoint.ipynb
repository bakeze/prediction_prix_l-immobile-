## 5. Préparation des données pour le modèle

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## 6. Création du modèle

```python
# Construction du modèle de Machine Learning 
model = LinearRegression()

# Entrainement du modèle sur les données 
model.fit(X_train, y_train)
```

```python
# print the intercept
print(model.intercept_)
print(model.coef_)
```

- L'intercept (ou constante) du modèle représente la valeur de la variable cible (y) lorsque toutes les variables indépendantes (X) sont égales à zéro. C'est un paramètre important car il permet de comprendre le point de départ de la prédiction lorsque toutes les caractéristiques sont nulles.

- Les coefficients de la régression linéaire indiquent l'importance de chaque caractéristique dans la prédiction de la variable cible. Chaque coefficient représente la variation de la variable cible pour une unité de variation de la caractéristique correspondante, toutes choses égales par ailleurs. Ces valeurs permettent de comprendre l'influence de chaque caractéristique sur les prédictions du modèle.

```python
coeff_df = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
y_predict = model.predict(X_test)
```

```python
plt.scatter(y_test,y_predict)
```

Interprétation des coefficients
Les coefficients représentent le changement dans la variable cible (prix des maisons, dans ce cas) pour chaque unité de changement dans la caractéristique respective, en supposant que toutes les autres caractéristiques restent constantes. Par exemple :

Une augmentation d'une unité dans "Revenu moyen de la zone" est associée à une augmentation de 21,62 $ dans le prix de la maison.

Une augmentation d'une unité dans "Âge moyen des maisons de la zone" est associée à une augmentation de 165,221,12 $ dans le prix de la maison, et ainsi de suite.
